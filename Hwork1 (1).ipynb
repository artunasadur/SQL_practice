```python
%%sql
CREATE TABLE IF NOT EXISTS fe512db.COMPUSTAT(
gvkey varchar(8),
datadate date,
fyearq int,
fqtr int,
fyr int,
tic varchar(8),
cusip varchar(11),
conm  varchar(30),
atq int ,
cshoq int,
ltq int,
cik int,
prccq int
);
```

```python
%%sql
LOAD DATA INFILE '/home/data/menagerie-db/compustat.csv' INTO TABLE fe512db.COMPUSTAT
FIELDS TERMINATED BY ','
OPTIONALLY ENCLOSED BY '"'
LINES TERMINATED BY '\n'
IGNORE 1 LINES
(@gvkey,@datadate,@fyearq,@fqtr,@fyr,@tic,@cusip,@conm,@atq,@cshoq,@ltq,@cik,@prccq)
SET
    gvkey = @gvkey,
    datadate = @datadate,
    fyearq = @fyearq,
    fqtr = @fqtr,
    fyr = @fyr,
    tic =  NULLIF(@tic, ''),
    cusip = NULLIF(@cusip, ''),
    conm = NULLIF(@conm, ''),
    atq = NULLIF(@atq, ''),
    cshoq = NULLIF(@cshoq, ''),
    ltq = NULLIF(@ltq, ''),
    cik = NULLIF(@cik, ''),
    prccq = NULLIF(@prccq, '')
    ;
#(@gvkey,@datadate,@fyearq,@fqtr,@fyr,@tic,@cusip,@conm,@atq,@cshoq,@ltq,@cik,@prccq)
```

```python
%%sql
ALTER TABLE fe512db.COMPUSTAT
ADD COLUMN debt_to_asset_ratio DOUBLE AS (ltq/atq) ;
```

Get the 3 stock tickers (tic) who has the largest av-
erage total assets (atq) in fyearq 2016. (Hint: you
can use order and limit to perform this result).

```python
%%sql
select tic,fyearq,atq
from fe512db.COMPUSTAT
where fyearq =2016
order by atq desc
limit 3;
```

List all company name and the value of outstanding
shares (cshoq) when the largest outstanding shares
between 2015 to 2017 is larger than 500.

```python
%%sql
SELECT conm, cshoq
FROM fe512db.COMPUSTAT
where fyearq between 2015 and 2017 and cshoq>500
;
```
